```python
import pandas as pd
from sqlalchemy import create_engine, text
import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
sakila = "sakila"
connection_string = 'mysql+pymysql://root:' + password + '@localhost/'+ sakila
engine = create_engine(connection_string)
engine
```

```python
def rentals_month(engine, month, year):
    with engine.connect() as connection:
        txt = f"""
                    SELECT customer_id, COUNT(rental_id) as rentals_{month}_{year}
                    FROM rental
                    WHERE MONTH(rental_date) = {month} AND YEAR(rental_date) = {year}
                    GROUP BY customer_id
                """
        query = text(txt)
        result = connection.execute(query)
        return pd.DataFrame(result.all())
```

```python
May_df = rentals_month(engine,5,2005)
May_df
```

```python
June_df = rentals_month(engine,6,2005)
June_df
```

```python
def rental_count_month(df, month, year):
    df.columns = [f'rentals_{month}_{year}' if col == 'rentals' else col for col in df.columns]
    return df
```

```python
May_renamed = rental_count_month(May_df,5,2005)
May_renamed
```

```python
June_renamed = rental_count_month(June_df,6,2005)
June_renamed
```

```python
def compare_rentals(df1, df2):
    combined_df = pd.merge(df1, df2, on='customer_id', how='outer')
    combined_df['difference'] = combined_df.iloc[:, 1] - combined_df.iloc[:, 2]
    return combined_df
```

```python
May_comparison = compare_rentals(May_df, May_renamed)
May_comparison
```

```python
June_comparison = compare_rentals(June_df, June_renamed)
June_comparison
```
